```python
class AIAgent:

    def __init__(self):
        self.owner = "Muhammad"
        self.battery = 80

    def start(self):
        print("========== AI CALL AGENT ==========")

        caller = input("Caller Name: ")
        relation = input("Relationship (family/friend/boss/unknown): ").lower()
        emergency = input("Emergency? (yes/no): ").lower()
        meeting = input("Is owner in meeting? (yes/no): ").lower()

        self.handle_call(caller, relation, emergency, meeting)

    def handle_call(self, caller, relation, emergency, meeting):

        if emergency == "yes":
            print("\nDecision:")
            print("Answering call immediately.")
            print("Sending alert to owner.")
            return

        if relation == "family":
            print("\nDecision:")
            print("Family call detected.")
            print("Connecting the call.")
            return

        if relation == "boss":
            print("\nDecision:")
            print("Boss is calling.")
            print("Connecting immediately.")
            return

        if meeting == "yes":
            print("\nDecision:")
            print("Owner is busy.")
            print("Putting call on hold.")
            print("Sending notification to owner.")
            return

        if relation == "friend":
            print("\nDecision:")
            print("Friend detected.")
            print("Ringing owner.")
            return

        print("\nDecision:")
        print("Unknown caller.")
        print("Ask caller for purpose.")
        print("Record voice message.")
        print("End call politely.")

agent = AIAgent()
agent.start()
```
